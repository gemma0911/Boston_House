#### TÍNH NĂNG/BIẾN ĐỘC LẬP trong Tập dữ liệu

1. CRIM:- (Crime) Tỷ lệ tội phạm trên mỗi dân cư theo thị trấn.

2. ZN:- (Zone) Tỷ lệ đất ở được quy hoạch cho các khu đất trên 25,000 feet vuông (khoảng 2.3 nghìn mét vuông).

3. INDUS:- (Industry) Tỷ lệ diện tích thương mại cho mỗi thị trấn.

4. CHAS:- (Charles River) Biến giả, 1 nếu nằm bên sông, 0 nếu không.

5. NOX:- (Nitric Oxides)  Nồng độ oxit nitric (NOX) trong không khí.

6. RM: (Rooms) Số phòng trung bình mỗi căn nhà.

7. AGE: (Age) Tỷ lệ căn nhà được xây trước năm 1940 và hiện vẫn còn giữ lại.

8. DIS: (distance) Khoảng cách trung bình đến các trung tâm việc làm tư nhân.

9. RAD: (Radial) Chỉ số khả năng tiếp cận đường cao tốc.

10. TAX: (Taxation) Mức thuế suất của ngôi nhà.

11. PTRATIO: (Pupil-teacher ratio) Tỉ lệ học sinh giáo viên (giáo viên tương ứng với học sinh) trong các trường học.

12. B: (Black)  trong đó: Bk là tỷ lệ người Mỹ gốc Phi theo thị trấn.

13. LSTAT: (Lower status) Tỷ lệ dân số có địa vị thấp.

# Z-score method

```python
#find the limits
upper_limit = df_stand['MEDV'].mean()+3*df_stand['MEDV'].std()
lower_limit = df_stand['MEDV'].mean()-3*df_stand['MEDV'].std()

print("upper limit: ", upper_limit)
print("lower limit: ", lower_limit)
```

```python
# Find the outliers
df_stand.loc[(df_stand['MEDV'] > upper_limit) | (df_stand['MEDV'] < lower_limit)]
```

Không có ngoại lệ trong cột MEDV trong Tập dữ liệu được chuẩn hóa khi sử dụng phương pháp điểm Z

# IQR method

```python
q1 = df_stand['MEDV'].quantile(0.25)

q3 = df_stand['MEDV'].quantile(0.75)

iqr = q3-q1
```

```python
q1, q3, iqr
```

```python
upper_limit = q3 + (1.5*iqr)
lower_limit = q1 - (1.5*iqr)

lower_limit, upper_limit
```

```python
sns.boxplot(x=df_stand['MEDV'])
```

```python
#Find outliers
df_stand.loc[(df_stand['MEDV'] > upper_limit) | (df_stand['MEDV'] < lower_limit)]
```

```python
##Cắt tỉa -xóa dữ liệu ngoại lệ

new_df = df_stand.loc[(df_stand['MEDV'] < upper_limit) & (df_stand['MEDV'] > lower_limit)]
print('before removing outliers: ', len(df_stand))
print('after removing outliers: ', len(new_df))
print('outliers: ', len(df_stand)-len(new_df))
```

```python
sns.boxplot(x=new_df['MEDV'])
```

```python
# Giới hạn - thay đổi giá trị ngoại lệ thành giá trị giới hạn trên (hoặc) dưới

new_df = df_stand.copy()
new_df.loc[(new_df['MEDV']>upper_limit),'MEDV'] = upper_limit
new_df.loc[(new_df['MEDV']<lower_limit),'MEDV'] = lower_limit
```

# 5. Xây dựng mô hình hồi quy tuyến tính đơn biến
### 1. Xác định biến mục tiêu trong tập dữ liệu của bạn.
  MEDV là biến mục tiêu trong bộ dữ liệu Giá nhà ở Boston.
  
### 2. Tách biệt các đặc điểm (X) và biến mục tiêu (y)

# Chia dữ liệu

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
# Chia dữ liệu thành 75% cho huấn luyện và 25% cho kiểm tra
X_train, X_test, y_train, y_test = train_test_split(new_df['AGE'], new_df['MEDV'], test_size=0.25, random_state=42)
```

```python
y_train
```

# Không dùng thư viện

```python
x_tb_age_train = X_train.mean()
y_tb_medv_train = y_train.mean()
```

```python
x_tb_age_train
```

```python
y_tb_medv_train
```

```python
sum_sq_diff_x_train = np.sum((X_train - x_tb_age_train) ** 2, axis=0)
sum_prod_diff_xy_train = np.sum((X_train - x_tb_age_train) * (y_train - y_tb_medv_train), axis=0)
```

```python
sum_sq_diff_x_train
```

```python
sum_prod_diff_xy_train
```

```python
# Tính hệ số góc (beta_1)
beta_1_train = sum_prod_diff_xy_train / sum_sq_diff_x_train
```

```python
beta_1_train
```

```python
# Tính hằng số chặn (beta_0)
beta_0_train = y_tb_medv_train - beta_1_train * x_tb_age_train
```

```python
beta_0_train
```

```python
# công thức : y = B0 + B1x + e
```

```python
def predict_y(x, beta_0, beta_1):
    return beta_0 + beta_1 * x
```

```python
# Dữ liệu mới để dự đoán
x_new = X_test
# Dự đoán giá trị y cho dữ liệu mới sử dụng hàm predict_y
y_pred_new = [predict_y(x, beta_0_train, beta_1_train) for x in x_new]
# In kết quả dự đoán
for i in range(len(x_new)):
    print("Dự đoán cho x =", x_new.iloc[i], "là y =", y_pred_new[i])
```

```python
import matplotlib.pyplot as plt

# Vẽ biểu đồ so sánh giữa giá trị thực tế và giá trị dự đoán
plt.figure(figsize=(20, 6))
plt.plot(y_test.values, label='Thực tế', marker='o', linestyle='-', color='blue')
plt.plot(y_pred_new, label='Dự đoán', marker='x', linestyle='-', color='red')
plt.xlabel('Mẫu')
plt.ylabel('Giá trị')
plt.title('So sánh giá trị thực tế và dự đoán trên tập kiểm tra')
plt.legend()
plt.grid(True)
plt.show()
```
